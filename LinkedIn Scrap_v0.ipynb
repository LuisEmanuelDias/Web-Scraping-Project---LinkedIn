```python
from bs4 import BeautifulSoup
import requests
website = 'https://www.linkedin.com/jobs/view/3075047477/'
response = requests.get(website)
content = response.text
soup = BeautifulSoup(content, 'lxml')

list_start_scrap = []
list_no_start_words = []
list_finish_scrap = []
list_garbage_words = []
list_technical_words = []

inicio = False
for x in re.findall('( |\n|a-zA-z)[a-zA-z]+( |\n)',soup.prettify()):
    x= x.strip().lower()
    if x in list_start_scrap:               #Si la palabra ya es una palabra de inicio i.e. "Requisitos" inicia la seccion de filtrado
        inicio = True
        continue
    if x in list_no_start_words: continue
    
    if not inicio:        #Determina si es una palabra de inicio
        respuesta = input(f"¿Esta palabra '{x}' marca el inicio? s/n").lower
        if respuesta == "s":
            inicio = True
            list_start_scrap += [x]
            continue
        list_no_start_words += [x]
        continue
    
    if x in list_finish_scrap: break
    
    if x not in list_garbage_words or x not in list_technical_words:
        respueta = (f"¿Esta palabra '{x}' es una skill técnica? s/n (si es el final de la seccion de requisitos coloque f)").lower
        if respuesta== 's':
            list_technical_words += [x]
            continue
        elif respuesta=='f':
            list_finish_scrap += [x]
            break
        list_garbage_words += [x]

with open('no_start_words',"a") as file:
    for word in list_no_start_words:
        file.write(word+"\n")
with open('start_words') as file:
    for word in list_start_scrap:
        file.write(word+"\n")
with open('finish_words') as file:
    for word in list_finish_scrap:
        file.write(word+"\n")
with open('garbage_words') as file:
    for word in list_garbage_words:
        file.write(word+"\n")
with open('technical_words') as file:
    for word in list_technical_words:
        file.write(word+"\n")
        
        
```
